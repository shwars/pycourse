# Блок 5: Вложенные циклы

Давайте начнем с повторения упражнения с последнего раздела. Нам нужно было:

1. Написать функцию `fact`, которая будет вычислять факториал $$n! = 1\cdot2\cdot3\cdot\dots\cdot n$$
2. Распечатать таблицу факториалов от 1 до 7

Давайте начнем с функции:

````python
def fact(n):
    acc = 1
    for i in range(2,n+1):
        acc = acc*i
    return acc

fact(7)
````

Здесь мы используем специальную технику: мы вводим переменную под названием **аккумулятор**, которая изначально принимает значение 1, а затем умножается на 2, 3 и т.д. до $n$. Это происходит внутри цикла. Когда цикл завершен - мы возвращаем значение накопителя, которое равно факториалу.

Число, которое изменяется от 2 до $n$ внутри цикла, называется **счетчиком цикла**. Это переменная, которая используется в выражении `for`, когда мы говорим `for i in ...`. Очень часто программисты называют эти переменные `i`,` j`, `k` и т.д., Но это просто традиция. Скорее всего, это потому, что `i` обозначает **index** или **integer** (а счетчики циклов часто являются целыми числами).

Здесь важно отметить, что в нашем случае нам необходимо изменить значение счетчика цикла с 2 на $n$. Если мы напишем для цикла таким образом:

```python
for i in range(n):
    ...
```

тогда `i` изменится с 0 на $n-1$, и цикл будет повторяться $n$ раз. В нашем случае мы хотим, чтобы значения `i` начинались с 2 и заканчивались значением $n$ включительно, поэтому нам нужно написать цикл примерно так:

```python
for i in range(2,n+1):
    ...
```

Важные вещи для запоминания:

* Первый аргумент `range` указывает начальное значение счетчика. Если аргумент отсутствует - по умолчанию он равен 0
* Второй аргумент `range` указывает, когда нам нужно завершить цикл, и цикл заканчивается непосредственно перед тем, как счетчик достигнет указанного значения. Так что это, по сути, последнее значение счетчика плюс 1

Теперь, чтобы распечатать таблицу факторных значений, мы используем другой цикл:

````python
for i in range(1,8):
    print(f'{i}! = {fact(i)}')
````

Обратите внимание, как элегантно мы используем здесь интерполяцию строк, помещая выражение `fact (i)` прямо в фигурные скобки внутри строки. Нам даже не нужно использовать дополнительную переменную!

## Вложенные циклы

В нашем случае для печати таблицы мы использовали цикл от 1 до 7, а внутри него для вычисления факториала был использован другой цикл внутри функции `fact`. Если бы мы писали все внутри одного большого куска кода, мы бы написали что-то вроде этого:

````python
for i in range(1,8):
    acc = 1
    for j in range(2,i+1):
        acc *= j
    print(f'{i}! = {acc}')
````

Когда у нас есть цикл внутри другого цикла, мы называем его **вложенным циклом**. Здесь стоит отметить две важные вещи:

* Нам нужно использовать разные имена для именования переменных счетчика во вложенных циклах. В нашем случае мы используем `i` в качестве счетчика ** внешнего цикла **, а` j` в качестве счетчика ** внутреннего цикла **.
* Счетчик внешнего цикла может использоваться внутри выражения диапазона во внутреннем цикле.

Мы также вводим здесь другую операцию, `* =`, которая очень полезна, когда мы имеем дело с аккумулятором. Выражение `acc * = j` в нашей программе эквивалентно` acc = acc * j`, но оно короче и все еще достаточно ясно с точки зрения смысла.

## Рисуем прямоугольный город

В предыдущем разделе мы узнали, как нарисовать сцену из нескольких домов. Теперь пришло время использовать понятие вложенных циклов, чтобы нарисовать прямоугольный город с множеством улиц! Нам нужно помнить, что после рисования одной улицы нам нужно переместить черепаху назад к началу улицы и вниз. Мы также определим конфигурацию города (количество строк и столбцов) в переменных в нашем коде - это всегда полезно чтобы сделать код более настраиваемым.

````python
import jturtle as turtle
````

````python
rows = 5 # число улиц
cols = 5 # число домов на уаждой улице
sep = 5 # расстояние между домами
w = 10 # размер домов

from pycourse import house, square
        
for _ in range(rows):
    # нарисуем одну улицу
    for _ in range(cols):
        house(w) # нарисуем дом
        # двигаем черепаху к следуйщему дому
        turtle.penup()
        turtle.right(90)
        turtle.forward(w+sep)
        turtle.left(90)
        turtle.pendown()
    # двигаем черепаху к началу улицы
    turtle.left(90)
    turtle.penup()
    turtle.forward(cols*(w+sep)) # ширина всей улицы
    turtle.right(90)
    turtle.backward(2*w+sep)
    turtle.pendown()
        
turtle.done()
````

Обратите внимание, что мы используем комментарии `#` довольно широко, чтобы помочь вам понять, что происходит внутри кода. Комментарии помогают другим людям, так же как и вам, через несколько недель понять код.

Другой полезный прием, который мы здесь используем, называется **анонимная переменная** `_` Если мы не заботимся о значении счетчика цикла, мы можем использовать подчеркивание вместо имени переменной, это означает, что *нам не важно ее значение*. Это помогает людям, которые читают код, чтобы понять, что значение переменной не важно, а также избавляет вас от необходимости каждый раз придумывать новое имя счетчика циклов внутри вложенных циклов.

## Вложенные циклы против одного цикла

Одну и ту же логику можно переписать, используя один большой цикл вместо двух вложенных. В нашем случае, когда мы сталкиваемся с концом улицы, нам нужно вернуть черепаху в начало улицы и опустить ее вниз. Предположим, что переменная «i» обозначает счетчик цикла, то есть номер дома, и в этом случае конец улицы встречается, когда это число делится на «cols», то есть остаток от «i», деленный на «cols», равен 0.

````python
for i in range(1,rows*cols+1):
    house(w) # draw a house
    # move turtle to the next house
    turtle.penup()
    turtle.right(90)
    turtle.forward(w+sep)
    turtle.left(90)
    turtle.pendown()
    if i%cols == 0: # end-of-street encountered
        # move turtle to the beginning of a street
        turtle.left(90)
        turtle.penup()
        turtle.forward(cols*(w+sep)) # width of the whole street
        turtle.right(90)
        turtle.backward(2*w+sep)
        turtle.pendown()
        
turtle.done()
````

Этот пример показывает нам, что всегда можно написать одну и ту же программу разными способами. И часто есть разные критерии, чтобы определить, какой путь лучше:

* **ясность кода** - всегда лучше иметь четкий и читаемый код, потому что его легче писать и поддерживать
* **эффективность программы** - иногда более простой или короткий код не означает повышения производительности, поэтому мы должны принимать во внимание соображения производительности

После этого нам приходится идти на компромисс между простотой и производительностью.

## Sierpiński Triangle

Now that we know loops, let me show you a very beautiful example of how a simple computer program can generate infinite complexity and beauty. We will learn how to draw so-called [**Sierpiński Triangle**](https://en.wikipedia.org/wiki/Sierpiński_triangle) - a fractal structure that looks like this:

<img src="images/Sierpinski_triangle.png" width="30%"/>

([photo](https://commons.wikimedia.org/w/index.php?curid=8862246) from [Wikipedia](https://en.wikipedia.org/wiki/Sierpiński_triangle) by Beojan Stanislaus, CC BY-SA 3.0)

This figure is surprisingly easy to draw! We need to implement the following algorithm:

1. Select three vertices of a triangle, $(x_1,y_1)$, $(x_2,y_2)$ and $(x_3,y_3)$.
2. Set coordinates of current point $(x,y)$ to some initial value.
3. Randomly pick one of the vertices $i\in\{1,2,3\}$
4. Calculate new point coordinates as *midpoint* between current point and selected vertice: 
$$x \leftarrow \frac{x_i+x}{2}$$
$$y \leftarrow \frac{y_i+x}{2}$$
5. Draw a point at coordinates $(x,y)$
6. Repeat from step 3

This process sounds like a loop than never ends, or so-called **infinite loop**. In practice, we definitely want our program to stop at some point, so we just program sufficiently large for-loop.

To begin with, let's define the list of triangle vertices as a list of coordinate pairs. We have not discussed yet what lists are, but for now please consider it just a collection of pairs, from which we would be able to pick one pair using `random.choice` function which we have seen before:

````python
vertices = [(0,00),(100,0),(50,100)]
````

Now let's program the main loop. Here we use coordinate-based graphics and `turtle.point` to draw the dot, and to make the dot smaller we specify additional parameter, `pointstyle`. Please note that this loop can take a moment to execute!

````python
import random

x,y = 150,150
for _ in range(3000):
    vx, vy = random.choice(vertices)
    x = (x+vx)/2
    y = (y+vy)/2
    turtle.point((x,y),pointstyle=',')
turtle.done()
````

Isn't this picture beautiful?! I find it very fascinating that we can achieve potentially infinite complexity with quite a short code.

While the figure is beautiful, our code can still be improved. In our code we think in terms of individual coordinates $x$ and $y$, while we can switch to higher level thinking, and keep in mind the points, and not coordinates. 

To do that, let's first define a function `midpoint` to calculate the position of middle point between points `a` and `b`. Let me remind you that if `a` denotes a pair of point coordinates, then `a[0]` denotes the first coordinate ($x$), and `a[1]` -- second coordinate ($y$).

Using this function, the code becomes a lot cleaner:

````python
def midpoint(a,b):
    return ((a[0]+b[0])/2,(a[1]+b[1])/2)

p = 150,150
for _ in range(3000):
    v = random.choice(vertices)
    p = midpoint(p,v)
    turtle.point(p, pointstyle=',')
turtle.done()
````
